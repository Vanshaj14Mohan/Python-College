## For Line Chart

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]
plt.plot(x, y)
plt.title("Line chart")
plt.show()
```

## For Bar chart

```python
plt.bar(["A", "B", "C"], [10,20,30])
plt.title("Bar chart")
plt.show()
```

## For Pie Chart

```python
labels = ["A", "B", "C"]
sizes = [50,30,20]
plt.pie(sizes, labels = labels)
plt.title("Pie chart")
plt.show()
```

## For Scatter Plot 

```python
x= [1,2,3,4,5]
y = [10,20,30,15,25]
plt.scatter(x,y)
plt.title("Scatter Plot")
plt.show()
```

## For Histogram


```python
data = [1,2,3,4,5,6,7]
plt.hist(data, bins =5)
plt.title("Histogram")
plt.show()
```

## For Box Plot:


```python
data = [1,2,2,3,3,3,4,5]
sns.boxplot(data = data)
plt.title("Box Plot")
plt.show()
```

## For Heatmap:

```python
data = np.random.rand(10,10)
sns.heatmap(data)
plt.title("Heatmap")
plt.show()
```

## For Area Chart:

```python
plt.fill_between([1,2,3,4,5], [1,4,5,6,7])
plt.title("Area Chart")
plt.show()
```

## For Pair Plot:


```python
df = sns.load_dataset("iris")
sns.pairplot(df)
plt.show()
```

## For Violin Plot:

```python
sns.violinplot(data = [1,2,3,4,5])
plt.title("Violin Plot")
plt.show()
```

## For Bubble Chart:

```python
sizes = [100,200,300,400,500]
plt.scatter(x,y, s = sizes)
plt.title("Bubble Chart")
plt.show()
```

## For 3D Plot:

```python
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure()
ax = fig.add_subplot(111,projection="3d")
x =[1,2,3,4,5]
y = [10,20,15,25,30]
z =[1,4,9,16,25]
ax.plot(x,y,z)
plt.title("3D Plot")
plt.show()
```
