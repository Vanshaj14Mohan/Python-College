```python
data = {"name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [20,22,21,23],
    "Grade": [85,90,78,92]}
df = pd.DataFrame(data)
print("Original DataFrame: ")
print(df)
```

```python
#Calculate mean age of students
mean_age = df["Age"].mean()
print("Average age is", mean_age)
```

```python
#For max age 
max_age = df["Age"].max()
print("Max age is", max_age)
```

```python
#For max and min grade
print("For maximum marks")
max_grade = df["Grade"].max()
print("Maximum grade is:", max_grade)
print("------------------")
print("For minimum marks")
min_marks = df["Grade"].min()
print("Minimum grade is:", min_marks)
```

```python
data = {"Student": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Math": [85,78,92,88,76],
    "English": [92,82,87,91,84],
    "Science": [88,79,85,93,80]}

df = pd.DataFrame(data)
print(df)
```

```python
threshold = 85
subject = "Math"

filtered_df = df[df[subject]>threshold]
print("\nFiltered Dataframe(Students who scored above", threshold, "in", subject+ "):")
print(filtered_df)
```

```python
#For bubble chart 
import matplotlib.pyplot as plt
x = [10,20,30,40,50]
y = [15,25,35,45,55]
sizes = [200,400,600,800,1000]
plt.scatter(x,y, s=sizes, alpha=0.5,c="blue")
for i in range(len(x)):
    plt.text(x[i], y[i], f"({x[i]}, {y[i]})", ha="center", va="center", color="blue")
plt.title("Bubble sort with lables")
plt.xlabel("X values")
plt.ylabel("Y values")
```

```python
import matplotlib.pyplot as plt 
import numpy as np
np.random.seed(0)
num_points = 10
x = np.random.rand(num_points)*100
y = np.random.rand(num_points)*100
sizes = np.random.rand(num_points)*1000
plt.figure(figsize=(10,6))
plt.scatter(x,y, s=sizes, alpha=0.7, edgecolors="w",linewidth=0.8)
plt.title("Basic bubble chart with random data")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```
