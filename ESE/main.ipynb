```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40, 50]
y = [15, 25, 35, 45, 55]
sizes = [100, 200, 300, 400, 500]  # Bubble sizes (area in points^2)
labels = [f'({xi}, {yi})' for xi, yi in zip(x, y)]  # Labels with x and y values

# Create the plot
fig, ax = plt.subplots()

# Plot the bubble chart
scatter = ax.scatter(x, y, s=sizes, alpha=0.5, c='pink', edgecolors='w', linewidth=2)

#Add labels to each bubble
for (xi, yi, label) in zip(x, y, labels):
    ax.text(xi, yi, label, fontsize=9, ha='center', va='center')

# Set labels and title
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_title('Bubble Chart with Labels')

# Show the plot
plt.show()
```

```python
import pandas as pd

# Create a sample dataframe of students' scores
data = {' Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Math': [90, 80, 75, 90, 80],
        'Science': [87, 92, 94, 85, 75],
        'English': [90, 91, 82, 87, 86]}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Set the threshold and subject
threshold = 86
subject = 'Math'

# Filter out the students who scored above the threshold in the specific subject
filtered_df = df[df[subject] > threshold]

print(f"\nStudents who scored above {threshold} in {subject}:")
print(filtered_df)
```
